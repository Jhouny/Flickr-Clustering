## Implements a map visualization of the dataset using folium library

```python
import pandas as pd
import numpy as np
import folium
```

### Load data

```python
data = pd.read_csv('../data/data_cleaned_titles.csv')
locations = data[["id", "lat", "long"]]
```

### Find the approximate mean location of all points to center the map

```python
map_center = [np.mean(locations["lat"]), np.mean(locations["long"])]
m = folium.Map(location=map_center, zoom_start=10)
```

### Add markers to the map

```python
from folium.plugins import FastMarkerCluster

fast_marker_cluster = FastMarkerCluster(
    location=map_center,
    tiles="Clustered data points",
    zoom_start=1,
    data=locations[["lat", "long"]].values.tolist(),
)

fast_marker_cluster.add_to(m)

folium.LayerControl().add_to(m)
```

```python
sw = locations[["lat", "long"]].min().values.tolist()
ne = locations[["lat", "long"]].max().values.tolist()

m.fit_bounds([sw, ne])
```
