## Crawler results processing

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/flickr_photo_urls.csv")
print(f"Total entries before cleaning: {len(df)}")
df.head()
```

>### Count null URL entries

```python
# Count null URL entries
null_url_count = df['photo_url'].isnull().sum()
print(f"Number of null URL entries: {null_url_count} - {null_url_count / len(df) * 100:.2f}% of total")
```

>### Remove duplicate entries

```python
# Remove duplicate entries by user and id fields
before_dedup = len(df)
df_cleaned = df.drop_duplicates(subset=['user', 'id'])
df_cleaned.reset_index(drop=True, inplace=True)
after_dedup = len(df_cleaned)
print(f"Number of entries after removing duplicates: {after_dedup} (removed {before_dedup - after_dedup})")
```

>### Save cleaned data

```python
# Save cleaned data
df_cleaned.to_csv("../data/flickr_photo_urls_cleaned.csv", index=False)
```
