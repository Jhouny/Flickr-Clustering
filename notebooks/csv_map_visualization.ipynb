## Implements a map visualization of the dataset using folium library

```python
import pandas as pd
import numpy as np
import folium
```

### Load data

```python
data = pd.read_csv('../data/data_cleaned_titles.csv')
locations = data[["id", "lat", "long"]]
```

```python
# Count empty location entries
empty_locations_count = locations.isnull().any(axis=1).sum()
print(empty_locations_count)
```

### Find the approximate mean location of all points to center the map

```python
map_center = [np.mean(locations["lat"]), np.mean(locations["long"])]
m = folium.Map(location=map_center, zoom_start=10)
```

### Add markers to the map

```python
for (lat, long, id) in zip(locations["lat"], locations["long"], locations["id"]):
    folium.Marker(
        location=[lat, long],
        popup=f"ID: {id}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)
```

```python
sw = locations[["lat", "long"]].min().values.tolist()
ne = locations[["lat", "long"]].max().values.tolist()

m.fit_bounds([sw, ne])
```
